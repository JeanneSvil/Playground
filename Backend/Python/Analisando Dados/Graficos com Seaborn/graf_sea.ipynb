### Importação

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Importação de Dados

```python
df_videos = pd.read_excel('videosYT.xlsx', 'videos')
display(df_videos)
```

### Criando um gráfico de dispersão:


```python
fig = sns.scatterplot(data=df_videos, x="Nº de Views", y="Nº de Likes", hue="Responsável",
                      style="Responsável", palette=["red", "black"])
plt.show(fig)
```

### Criando um gráfico de dispersão(relacional):


```python
grafico_rel = sns.relplot(data=df_videos, x="Nº de Views", y="Nº de Likes", hue="Categoria",
                          col="Responsável", )
grafico_rel.set_titles("Responsável: {col_name}")
plt.show(grafico_rel)
```

### Criando um gráfico de linha:

```python
df_incritos = pd.read_excel('videosYT.xlsx', 'Inscritos')
display(df_incritos)
```

```python
graf_linha = sns.lineplot(data= df_incritos, x= 'Mês/Ano', y='Inscritos', color= 'red')
plt.show(graf_linha)
```

### Criando Histogramas:

```python
fig = sns.displot(data=df_videos, x="Nº de Views")
```

```python
fig = sns.displot(data=df_videos, x="Nº de Views", kind= 'kde')
```

```python
fig = sns.displot(data=df_videos, x="Nº de Views", kind= 'ecdf')
```

```python
fig = sns.displot(data=df_videos, x="Nº de Views", hue= "Responsável", col= "Responsável", rug= True)
```

```python
fig = sns.pairplot(data=df_videos,hue= "Responsável")
```

### Categorical Plots:

```python
fig = sns.countplot(data=df_videos, x="Nº de Views", hue= "Responsável")
```

### Criando uma regressão linear:


```python
graf_regre = sns.regplot(data=df_videos, x="Nº de Views", y="Nº de Likes")
```

```python
graf_regre = sns.lmplot(data=df_videos, x="Nº de Views", y="Nº de Likes", hue= "Responsável",
                        markers= ['o', 'x'])
plt.show(graf_regre)
```
